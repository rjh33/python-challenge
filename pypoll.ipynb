```python
percentageList=[]
#The percentage of votes each candidate won
for i in candidates:     
    pctOfVotes= int(candidate.count(i))/numberOfVotes*100
    pctOfVotes= str(round(pctOfVotes,2))
    percentageList.append(pctOfVotes)
print(percentageList)
```

```python
#The total number of votes each candidate won
candidateVotes=[]
for j in candidates:
    totalNumberofVotes= candidate.count(j)
    candidateVotes.append(totalNumberofVotes)   
   
```

```python
#naming the winner
maxOfVotes = max(candidateVotes)
maxLocation=candidateVotes.index(maxOfVotes)       
winner = candidates[maxLocation]    

 
        
        
```
